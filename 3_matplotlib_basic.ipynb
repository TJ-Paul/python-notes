# import matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

# basic plotting

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

plt.plot(xs,ys)
plt.show()
```

## grid

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

plt.plot(xs,ys)
plt.grid(True)
plt.show()
```

## marker

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

plt.plot(xs,ys, marker = 'o')
plt.show()
```

## linestyle, color

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

plt.plot(xs,ys, marker = 'o', linestyle = '--', color = 'red')
plt.show()
```

# scatter

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

plt.scatter(xs,ys, color = 'red')
plt.show()
```

## scatter size

```python
plt.figure(figsize=(4,2))

xs = np.array([1,2,3,7,10])
ys = np.array([12,32,42,45,11])

sizes = np.array([4]*5) # creates a standard Python list with five repetitions of the number 2 : [2, 2, 2, 2, 2].
plt.scatter(xs,ys, color = 'red', s = sizes)
plt.show()
```

# bar

# subplot

```python
# --- Top Plot Data ---
xs1 = np.array([1, 2, 3, 7, 10])
ys1 = np.array([12, 32, 42, 45, 11])

plt.figure(figsize=(5, 9)) 
plt.title('first basic plotting')

# --- First Subplot (Top) ---
plt.subplot(2, 1, 1) # 2 rows, 1 column, 1st plot
plt.xlabel('x s')
plt.ylabel('y s')
plt.plot(xs1, ys1, marker='o', linestyle='--', color='red')

# --- Bottom Plot Data ---
xs2 = np.array([1, 2, 6, 7, 1])
ys2 = np.array([12, 32, 45, 32, 20]) # Added a 5th number (20) so the lengths match

# --- Second Subplot (Bottom) ---
plt.subplot(2, 1, 2) # 2 rows, 1 column, 2nd plot
plt.xlabel('x s')
plt.ylabel('y s')
plt.plot(xs2, ys2, marker='o', linestyle='--', color='green') # Added ys2 here

plt.show()
```

# linspace

## sine wave

```python
plt.figure(figsize=(5,3))
sampling_f = 1000
t = np.linspace(0,1,sampling_f)

f = 5
y = np.sin(2 * np.pi * f * t)

plt.plot(t,y)
plt.show()
```

## square wave

how np.sign() works:

> If a value is positive ($y > 0$), it becomes 1.0.

> If a value is negative ($y < 0$), it becomes -1.0.

> If a value is exactly zero, it stays 0.0 

```python
plt.figure(figsize=(5,3))
sampling_f = 1000
t = np.linspace(0,1,sampling_f)

f = 5
y = np.sin(2 * np.pi * f * t)

square_wave = np.sign(y)


plt.plot(t,square_wave)
plt.show()
```

## stem

```python
plt.figure(figsize=(5,3))
sampling_f = 1000
t = np.linspace(0,1,sampling_f)

f = 5
y = np.sin(2 * np.pi * f * t)

plt.stem(t,y)
plt.show()
```
