# Function

```python
def add(a,b):
    return a+b

print(add(5,3))
```

```python
def subtract(a,b,c=0):
    return a-b+c

print(subtract(4,3,2))
print(subtract(4,3))
```

```python
def math_ops(a,b):
    return a+b, a-b, a*b

x, y, z = math_ops(5,9)

print(x)
print(y)
print(z)
```

```python
def varfnc(a, *args):
    print(a)
    print(args)

varfnc(5,9,8,7,3)
```

# OOP

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def talk(self):
        return f'{self.name} talks'
    
    def __str__(self):
        return f'{self.name} and {self.age}'
```

```python
a = Animal('Tom', 6)
print(a)
```
